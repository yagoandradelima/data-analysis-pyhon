```python
import pandas as pd

df = pd.read_csv('../data/supermarket_sales.csv')

df.head()
```

### Adaptações

Para questões unicamente didáticas estou alterando o nome das cidades (coluna 'city') dentro do conjunto de dados original.

```python
# Analisando o tamanho do Dataset
# Quero realizar a alteração de valores na coluna city, se for muito grande aplico apply
df.shape
```

```python
# Dataset dentro dos padrões mínimos.
# Com poucos valores únicos, utilizei o replace para realizar a troca de valores

cities = list(df["City"].unique())

df["City"] = df["City"].replace({cities[0]:"Ananindeua", cities[1]:"Teresina", cities[2]:"Sao Luis"})

df
```

```python
# Checando se existem valores únicos para cada ID no dataset
df["Invoice ID"].nunique()
```

```python
df["Total value"] = df["Unit price"] * df["Quantity"]

df
```

```python
df.groupby("Product line").agg({"Total value":"sum",
                                "Quantity":"sum",
                                "Unit price":"mean"
                            }).rename(columns={"Unit price":"Preco unitario",
                                                "Total value":"Vendas totais",
                                                "Quantity":"Unidades vendidas"
                            }).sort_values(by=["Vendas totais", "Unidades vendidas", "Preco unitario"],
                                               ascending=[False, False, False]).reset_index()
```

```python
df.groupby("City").agg({"Total value":"sum",
                        "Quantity":"sum",
                        "Unit price":"mean"
                            }).rename(columns={"Unit price":"Preco unitario",
                                                "Total value":"Vendas totais",
                                                "Quantity":"Unidades vendidas"
                            }).sort_values(by=["Vendas totais", "Unidades vendidas", "Preco unitario"],
                                               ascending=[False, False, False]).reset_index()
```
